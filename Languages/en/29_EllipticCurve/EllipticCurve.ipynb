```python
import numpy as np
import matplotlib.pyplot as plt

# 定义椭圆曲线 y^2 = x^3 + ax + b，这里使用示例参数
a = -1
b = 1

# 生成椭圆曲线上的点
x = np.linspace(-2, 3, 400)
y = np.sqrt(x**3 + a*x + b)
y_neg = -y

# 绘制椭圆曲线
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'b')
plt.plot(x, y_neg, 'b')

# 点 P 和 Q: x_p != x_q
x_p = 0
x_q = 1
P = np.array([x_p, np.sqrt(x_p**3 + a*x_p + b)])
Q = np.array([x_q, np.sqrt(x_q**3 + a*x_q + b)])

# 计算 R
if np.array_equal(P, Q):
    # 点加倍情况
    lambda_ = (3 * P[0]**2 + a) / (2 * P[1])
else:
    # 普通情况
    lambda_ = (Q[1] - P[1]) / (Q[0] - P[0])

x3 = lambda_**2 - P[0] - Q[0]
y3 = lambda_ * (P[0] - x3) - P[1]
R = np.array([x3, -y3])  
R_dot = np.array([x3, y3]) # 取反射点以符合椭圆曲线的加法规则

# 计算经过 P 和 Q 的直线
line_x = np.linspace(-2, 2, 2)
line_y = lambda_ * (line_x - P[0]) + P[1]
plt.plot(line_x, line_y, 'r--', label='Line through P and Q')

# 绘制点
plt.plot(P[0], P[1], 'go', markersize=10, label='P')
plt.plot(Q[0], Q[1], 'ro', markersize=10, label='Q')
plt.plot(R[0], R[1], 'ko', markersize=10, label='R')
plt.plot(R_dot[0], R_dot[1], 'ko', markersize=10, label='R` (P+Q)')

# 添加注释
plt.annotate('P', xy=P, xytext=(P[0], P[1] + 10), textcoords='offset points')
plt.annotate('Q', xy=Q, xytext=(Q[0], Q[1] + 10), textcoords='offset points')
plt.annotate('R', xy=R, xytext=(R[0], R[1] + 10), textcoords='offset points')
plt.annotate('R` (P+Q)', xy=R_dot, xytext=(R_dot[0], R_dot[1] + 10), textcoords='offset points')

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Elliptic Curve Addition Examples')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
